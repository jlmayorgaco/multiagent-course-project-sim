```python
from mesa.visualization import SolaraViz, make_space_component

def agent_portrayal(agent):
    return {
        "shape": "circle",
        "color": "green" if hasattr(agent, "state") and agent.state == "natural" else "red",
        "r": 0.5,
        "filled": True,
    }

component = make_space_component(agent_portrayal)

model = ForestModel(width=10, height=10, num_palms=5, num_drones=2, num_people=1)

SolaraViz(model=model, components=[component])

```
